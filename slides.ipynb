# 1. 梯度优化

对于函数


```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

xaxis = np.arange(-1, 1, 0.01)
yaxis = np.arange(-1, 1, 0.01)
z1 = np.array([x ** 2 for x in xaxis])
z2 = np.array([y ** 2 for y in yaxis])
zaxis = z1 + z2
X, Y = np.meshgrid(xaxis, yaxis)
Z = X ** 2 + Y ** 2

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 100)
```

构造函数表达式和偏导数，
对 x， y 赋初值 (233，233)

```python
from sympy import *

x, y, z = symbols("x y z")
z = x ** 2 + y ** 2
dzdx = diff(z, x)
dzdy = diff(z, y)

xmin, ymin = 233, 233
alpha = 0.1
grad = np.array([dzdx.subs([(x, xmin), (y, ymin)]),
                 dzdy.subs([(x, xmin), (y, ymin)])],
                 dtype=np.float64)
```

执行 15 次。

```python
for i in range(15):
    grad = np.array([dzdx.subs([(x, xmin), (y, ymin)]),
                     dzdy.subs([(x, xmin), (y, ymin)])],
                     dtype=np.float64)
    xmin -= alpha * grad[0]
    ymin -= alpha * grad[1]
    
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(xaxis, yaxis, zaxis, 'gray')
ax.scatter3D([xmin], [ymin], y.subs([(x, xmin), (y, ymin)]))
```

执行 30 次。

```python
for i in range(15):
    grad = np.array([dzdx.subs([(x, xmin), (y, ymin)]),
                     dzdy.subs([(x, xmin), (y, ymin)])],
                     dtype=np.float64)
    xmin -= alpha * grad[0]
    ymin -= alpha * grad[1]
    
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(xaxis, yaxis, zaxis, 'gray')
ax.scatter3D([xmin], [ymin], y.subs([(x, xmin), (y, ymin)]))
```

执行 35 次。

```python
for i in range(5):
    grad = np.array([dzdx.subs([(x, xmin), (y, ymin)]),
                     dzdy.subs([(x, xmin), (y, ymin)])],
                     dtype=np.float64)
    xmin -= alpha * grad[0]
    ymin -= alpha * grad[1]
    
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(xaxis, yaxis, zaxis, 'gray')
ax.scatter3D([xmin], [ymin], y.subs([(x, xmin), (y, ymin)]))
```

直到收敛：


```python
while(np.linalg.norm(grad) != 0):
    grad = np.array([dzdx.subs([(x, xmin), (y, ymin)]),
                     dzdy.subs([(x, xmin), (y, ymin)])],
                     dtype=np.float64)
    xmin -= alpha * grad[0]
    ymin -= alpha * grad[1]
    
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(xaxis, yaxis, zaxis, 'gray')
ax.scatter3D([xmin], [ymin], y.subs([(x, xmin), (y, ymin)]))
ax.plot3D(xaxis, yaxis, 0)
```
